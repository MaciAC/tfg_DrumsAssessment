# MDB DRUMS

Get github urls of wav files and txt from MDB Drums dataset

 https://github.com/CarlSouthall/MDBDrums

```python
drum_urls_file = 'data/MDB-Drums/drum_files.txt'
drum_files = get_MusicDelta_urls(drum_urls_file)

annotations_urls_file = 'data/MDB-Drums/annotation_files.txt'
annotation_files = get_MusicDelta_urls(annotations_urls_file)
```

### Create song dictionary

Each key contains a list of annotations and the array containing the audio file

```python
legend = dict()
song_dict = dict()

for drum_file, annotation_file in zip(drum_files ,annotation_files):
    song = drum_file.split('/')[-1].split('_')[1]
    
    annotations = get_Annotations_list_from_url(annotation_file, legend, fs)
    
    song_dict[song] = { "audio" : ess.MonoLoader(filename = drum_file, sampleRate = fs)(),
                        "annotations" : annotations }
```
